```python
import pandas as pd
from google.colab import drive
import json
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from sklearn.preprocessing import LabelEncoder

drive.mount('/content/drive')
```

```python
df_train = pd.read_csv("/content/drive/MyDrive/product-recommendation-challenge/train.csv")
df_items = pd.read_csv("/content/drive/MyDrive/product-recommendation-challenge/item_metadata_filtered.csv", on_bad_lines='warn')
with open("/content/drive/MyDrive/product-recommendation-challenge/id_mappings.json", "r") as f:
    id_mappings = json.load(f)
```

```python
item_reverse_map = id_mappings['item_reverse_mapping']
df_train['parent_asin'] = df_train['item_id'].astype(str).map(item_reverse_map)
df = df_train.merge(df_items, how='inner', on='parent_asin')
```

```python
df.drop(columns=['user_id', 'item_id', 'parent_asin'], inplace=True)
```

```python
for col in ['main_category', 'store', 'category']:
    df[col] = df[col].fillna("unknown")
    df[col] = LabelEncoder().fit_transform(df[col])
```

```python
df['rating_number'] = np.log1p(df['rating_number'].fillna(0))
df['price'] = np.log1p(df['price'].fillna(0))
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
df[['average_rating', 'rating_number', 'price']] = scaler.fit_transform(
    df[['average_rating', 'rating_number', 'price']]
)
```
